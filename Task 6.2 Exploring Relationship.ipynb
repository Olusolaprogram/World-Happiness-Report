# 04 Creating a Scatterplot

```python
# Rename columns
df_happiness = df_happiness.rename(columns={'Economy (GDP per Capita)': 'Economy_GDP_per_Capita', 'Happiness Score': 'Happiness_Score'})
```

```python
# Create scatterplot
sns.lmplot(x='Economy_GDP_per_Capita', y='Happiness_Score', data=df_happiness)
# Show the plot
plt.show()
```

```python
# Family and happiness scatterplot
sns.lmplot(x = 'Family', y = 'Happiness_Score', data = df_happiness)
```

```python
# Life expectancy and happiness scatterplot
sns.lmplot(x = 'Health (Life Expectancy)', y = 'Happiness_Score', data = df_happiness)
```

# 08 creating categorical plot

```python
# Creating a histogram to see distribution
sns.histplot(df_happiness['Happiness_Score'], bins = 10, kde = True)
```

```python
# Creating score category for categorical plot
df_happiness.loc[df_happiness['Happiness_Score'] < 4.5 , 'Score category'] = 'Low score'
```

```python

df_happiness.loc[(df_happiness['Happiness_Score'] >= 4.5) & (df_happiness['Happiness_Score'] < 6.5), 'Score category'] = 'Middle score'
```

```python
df_happiness.loc[df_happiness['Happiness_Score'] >= 6.5, 'Score category'] = 'High Score'
```

```python
df_happiness['Score category'].value_counts(dropna = False)
```

```python
# Creating categorical plot with GDP per Capita
sns.set(style="ticks")
cat_plot_GDP = sns.catplot(x='Economy_GDP_per_Capita', y='Happiness_Score', hue="Score category", data=df_happiness)
```

```python
# Creating categorical plot with Family
sns.set(style="ticks")
cat_plot_family = sns.catplot(x='Family', y='Happiness_Score', hue="Score category", data=df_happiness)
```

```python
# Creating categorical plot with life expectancy
sns.set(style="ticks")
cat_plot_life = sns.catplot(x='Health (Life Expectancy)', y='Happiness_Score', hue="Score category", data=df_happiness)
```

```python
# Creating categorical plot with Freedom
sns.set(style="ticks")
cat_plot_freedom = sns.catplot(x='Freedom', y='Happiness_Score', hue="Score category", data=df_happiness)
```

```python

# Creating categorical plot with Gov_trust
sns.set(style="ticks")
cat_plot_gov = sns.catplot(x='Trust (Government Corruption)', y='Happiness_Score', hue="Score category", data=df_happiness)
```

```python
# Creating categorical plot with generosity
sns.set(style="ticks")
cat_plot_generosity = sns.catplot(x='Generosity', y='Happiness_Score', hue="Score category", data=df_happiness)
```
